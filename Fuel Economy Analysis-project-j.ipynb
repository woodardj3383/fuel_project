## Which model has the highest `air_pollution_score` and fuel economy in 2008 and 2018? Compare them. 

```python
airmax_08 = df_alpha_08.air_pollution_score.max()
print(airmax_08)

mileage_max_08 = df_alpha_08.cmb_mpg.max()
print(mileage_max_08)
```

```python
best_08 = df_alpha_08.query(f'air_pollution_score == {airmax_08} and cmb_mpg == {mileage_max_08}')
best_08
```

```python
df_alpha_08.query('air_pollution_score == 9')
```

```python
best_08_model = best_08.iloc[0].model
best_mpg_08 = best_08.iloc[:, 7:10].iloc[0]
best_mpg_08
```

```python
airmax_18 = df_alpha_18.air_pollution_score.max()
airmax_18
```

```python
mileage_max_18 = df_alpha_18.cmb_mpg.max()
mileage_max_18
```

```python
best_18 = df_alpha_18.query(f'air_pollution_score == {airmax_18} and cmb_mpg == {mileage_max_18}')
best_model_18 = best_18.iloc[0].model
best_mpg_18 = best_18.iloc[:, 7:10].iloc[0]
best_mpg_18
```

```python
cars = [f'{best_08_model} (2008)', f'{best_model_18} (2018)']

plt.subplots(figsize=(15, 12))
plt.bar(best_mpg_08.index, best_mpg_08, color='b', edgecolor='black', alpha=0.5)
plt.bar(best_mpg_18.index, best_mpg_18, color='g', edgecolor='black', alpha=0.5)
plt.title('(2008 vs 2018) Most Economic Car by Year', fontsize=16)
plt.ylabel('Fuel Economy', fontsize=14)
plt.xlabel('Categories (MPG)', fontsize=14)
plt.legend(cars, loc='best', prop={'size': 14});
```
